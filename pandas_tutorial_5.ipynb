# Pandas Tutorial 5: Handling Missing Data - fillna, interpolate, dropna

Topics covered in this tutorial:
* Fillna to fill missing values using different ways
* Interpolate to make a guess on missing values using interpolation
* Dropna to drop rows with missing values

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\Vaishob\\PycharmProjects\\pandas\\weather_data.csv",parse_dates=['day'])
df.set_index('day',inplace=True)
df                               
```

### `fillna` 
* Fill all NaN with one specific value

```python
new_df = df.fillna(0)
new_df
```

### Dictionary input to `fillna`
Often times it makes sense to replace these NaN values with something meaningful or a 'guess'

```python
new_df = df.fillna({
    'temperature': 0, 
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
# method called 'ffill' means forward fill
# carries forward the value from the previous row 
new_df = df.fillna(method="ffill")
new_df
```

```python
# method called 'bfill' means backwards fill
# copies the values from the next row
new_df = df.fillna(method="bfill")
new_df
```

### `axis`
* Can be either 'rows' or 'columns'
* Decides where the values are carried from

```python
# Does the method e.g. backwards fill, but in 
# horizontal axis (horizontal = columns)
new_df_1 = df.fillna(method="bfill", axis="columns")
new_df_1
```

 ### 'limit' param of `fillna`
 * Max number of consecutive NaN values to forward fill or backward fill

```python
new_df = df.fillna(method="ffill", limit=1)
new_df
```

### `interpolate()` method
* Fill NaN values using an interpolation method (default linear interpolation)

```python
new_df = df.interpolate()
new_df
```

### 'method' param of `interpolate`
* passing time as argument works on daily and higher resolution data to interpolate given length of interval

```python
new_df = df.interpolate(method="time")
new_df
```

### `dropna` 
* Remove missing data

```python
new_df = df.dropna()
new_df
```

### `how` parameter 
* How to work with missing data

```python
new_df = df.dropna(how="all")
new_df
```

### `threshold` parameter 
* If I have at least these many non-NaN values, keep the row

```python
new_df = df.dropna(thresh=2)
new_df
```

### Insert missing dates
* Insert missing values by creating a `date_range`
    * Reindex the dataframe using the `date_range` object

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
